```
# Import necessary libraries and modules
from skimage import io
import os
import numpy as np # linear algebra
import matplotlib.pyplot as plt # Data Visualization 
```

```
# Input compressed data
centers = np.load('compressedk=4.npy')
c_image = io.imread('compressedk=4.png')
```

```
# Reshape image again to make it 3D
image = np.zeros((c_image.shape[0],c_image.shape[1],3),dtype=np.uint8 )
```

```
# ReConstruct image
for i in range(c_image.shape[0]):
    for j in range(c_image.shape[1]):
            image[i,j,:] = centers[c_image[i,j],:]
```

```
# Plot the Compressed and the Decompressed image next to each other
fig, ax = plt.subplots(1, 2, figsize = (12, 8))

ax[0].imshow(c_image)
ax[0].set_title('Compressed Image')

ax[1].imshow(image)
ax[1].set_title('Reconstructed Image')

for ax in fig.axes:
    ax.axis('off')
plt.tight_layout();
```

# New Section

```
# Output image
io.imsave('reconstructedk=4.png',image);
```
